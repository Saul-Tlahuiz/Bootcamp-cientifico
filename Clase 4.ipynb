```julia
#=
Multiplicacion de Matrices 
Datos de entrada vectores
Datos de salida: Un numero real si los vectores son de la misma dimension
NaN si los vectores no son de la misma dimension
=#
#v1 * v2 = Σ(ai*bi)
function prodPunto(v1,v2)
    n1 = length(v1)
    n2 = length(v2)
    resultado = 0.0
    if n1 == n2
        for i=1:n1
            resultado = resultado + v1[i]*v2[i]
        end
    return resultado
    end
    return NaN
end
v1 = [1,-1,3]
v2 = [2,-3,1]
v3 = [1,1,1]
v4 = [0,0,1]
println(prodPunto(v3,v4))
```

```julia
#=
Funcion Matriz Vector
=#
function matrizVector(m,v)
    ren,col = size(m)
    n = length(v)
    resultado = zeros(ren)
    for j = 1:ren
        resultado[j] = prodPunto(m[j,:],v)
        #for i = 1:col
         #   resultado[j] = resultado[j] + m[j,i]*v[i]
        #end
    end
    return resultado
end
matriz = ones(3,4)
vector = ones(4)
println(matrizVector(matriz,vector)) 
```

```julia
#otra forma de hacer el calculo de matrix * vector
function matrizVector(m,v)
    ren,col = size(m)
    n = length(v)
    resultado = zeros(ren)
    for j = 1:ren,i=1:col #asi podemos poner 2 indices en el mismo for para ahorrarnos un for
        resultado[j] = resultado[j] + m[j,i]*v[i]
    end
    return resultado
end
matriz = ones(3,4)
vector = ones(4)
println(matrizVector(matriz,vector)) 
```

```julia
#=
Producto de Matrices
=#

function matrizMatriz(A,B)
    ra,ca = size(A)
    rb,cb = size(B)
    c = zeros(ra,cb)
    for i=1:ra
        for j=1:cb
            for k=1:ca
                c[i,j] = c[i,j] + A[i,k] * B[k,j] 
            end
        end
    end
    return c
end
matrizA = ones(2,2)
matrizB = ones(2,2)
println(matrizMatriz(matrizA,matrizB))
```
