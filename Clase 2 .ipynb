```julia

#vamos a crear nuestra funcion de valor absoluto
function absoluto(x)
    if x <= 0
        x = -x
    end
end
println(absoluto(4))
#=
La funcion que se va a programar sirve para calcular la raiz cuadrada de un numero real
funcion:raiz
Datos de entrada: x = real
Datos de salida: a = real
a = x
b = 1
mientras |a-b| > tolerancia (while)
  a = (a+b)/2
  b = x/a

println(raiz(4))
=#

function raizc(x)
    a = x
    b = 1
    tol = 0.01
    img = 1.0
    if x < 0
        img = 1im
        x = -x
    end
      while absoluto(a-b) > tol
         a = (a+b)/2
         b = x/a
      end
      return a*img
end
println(raizc(4))
#corregir esta funcion
```

```julia
function f(x)
    return 4/(1+x*x)
end
#la integral de esta funcion 4/(1+x*x) se aproxima a pi
#=
Nuevo algoritmo de integracion
Datos de entrada: A = real, B = real, f(x) = funcion,num_pasos = entero
Datos de salida: Resultado -> real
procedimiento o algoritmo:

=#
function integralNum(ini,fin,num_pasos)
    δ = (fin-ini)/num_pasos
    x = δ/2 + ini
    suma = 0
    for i = 1:num_pasos
        suma = suma + f(x)
        x = x+δ
    end
    resultado = suma*δ
    return resultado
end
println(integralNum(0.0,1.0,1000))
```

```julia
#=
Proposito: crearemos una funcion que nos ayuda a convertir grados Fahrenheit a Celsius con la siguiente
formula: C = 5/9F - 32°
Datos de Entrada 
F = real
Datos de salida
C = real
Ejemplo
fahrenheitToCelsius(0) = -17.777
celsiusAFahrenheit(200) = 392.0
=#
function fahrenheitToCelsius(F)
    C = 5/9 * (F -32)
    return C
end
fahrenheitToCelsius(90)

function celsiusAFahrenheit(C)
    F = 9/5*C + 32
    return F
end
#Funcion principal
function main()
    println("Escribe los grados que deseas convertir")
    grados = parse(Float64,readline())
    resultado = fahrenheitToCelsius(grados)
    resultado2 = celsiusAFahrenheit(grados)
    println("$grados ° Fahrenheit en grados Celsius son $resultado")
    println("$grados ° Celsius en grados Fahrenheit son $resultado2")
end
main()

#=
Toda interaccion del usuario con el codigo se hace en la funcion principal no en las otras funciones 
creadas, ademas cada funcion creada debe hacer solo una cosa, por ejemplo, si la funcion es para calcular
solo debe calcular,etc.
=#
```
